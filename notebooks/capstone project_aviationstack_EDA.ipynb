Send a request to AviationStack for the Historical Flights 

```python
# imports

import requests
import pandas as pd
from pprint import pprint
api_key = '07bb24d2632ebac61237a274382fdd0d'


# Define function to get historical flight info from AviationStack
def get_historical_flight(date_info, iata_code):

        # Define the API endpoint url and the parameters
        api_url = "https://api.aviationstack.com/v1/flights"

        params = {
                'access_key': api_key,
                'flight_date': date_info,
                'dep_iata': iata_code,
        }

        # Use the 'get' method from the requests library to fetch data from the API
        api_result = requests.get(api_url, params)

        # Print the status of request message.
        if api_result.status_code == 200:
                print('Request successful')
                api_response = api_result.json()
                return api_response
        
        else:
                print('Request NOT successful: ', api_result.status_code)
                return api_result.json()
```

Parse results and put same into a DataFrame.

```python
# Define function to parse results and put same in DataFrame
def parse_flight_data(data_from_api, dept_airport_city):

    airline_name = []
    flight_number = []
    flight_iata = []
    flight_date = []
    flight_destination = []
    departure_estimated = []
    departure_actual = []
    depature_delay = []
    departure_airport = []
    departure_airport_city = []

    for i in data_from_api['data']:
        airline_name.append(i['airline']['name'])
        flight_iata.append(i['airline']['iata'])
        flight_destination.append(i['arrival']['airport'])
        departure_actual.append(i['departure']['actual'])
        departure_airport.append(i['departure']['airport'])
        depature_delay.append(i['departure']['delay'])
        departure_estimated.append(i['departure']['estimated'])
        flight_number.append(i['flight']['number'])
        flight_date.append(i['flight_date'])
        departure_airport_city.append(dept_airport_city)
        
    
    # Parse list to Pandas Dataframe
    df_parsed_data = pd.DataFrame({
        "airline_name":airline_name, "flight_number":flight_number, "flight_iata":flight_iata, "flight_date": flight_date, 
        "flight_destination": flight_destination, "departure_estimated":departure_estimated, "departure_actual": departure_actual,
        "departure_delay": depature_delay, "departure_airport":departure_airport, "departure_airport_city": departure_airport_city })
    return df_parsed_data
    
```

Join the data from all 4 dataframes to create a new Dataframe.

```python
# Merge all 4 Dataframes to create a new master Dataframe using Concatenate method on the header column (axis = 0)

df_master_aviation_flight_data = pd.concat([
    df_aviation_flight_data_YYC, df_aviation_flight_data_YYZ, 
    df_aviation_flight_data_YVR, df_aviation_flight_data_YUL], axis=0, ignore_index=True, sort=False)

# Code below allows me to persistently store variables across different notebook sessions that I can retrieve later.
%store df_master_aviation_flight_data 

# Display Dataframe for Master Data (from all 4 airports - YYC, YYZ, YVR, YUL)
df_master_aviation_flight_data
```

Provide a visualization as part of my EDA process. 

```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# Group by 'Departure Airport' and get the average flight delay for the day
df_avg_delay = df_master_aviation_flight_data.groupby('departure_airport')['departure_delay'].mean()
df_avg_delay_plot_ready = df_avg_delay.reset_index()

# Create a histogram plot
plt.figure(figsize=(12, 6))
sns.barplot(x='departure_airport', y='departure_delay', data=df_avg_delay_plot_ready, palette='viridis')
plt.title('Average Flight Delay (in minutes) for each Canadian Airport')
plt.xlabel('departure_airport')
plt.ylabel('Average Delay (mins)')
plt.show()
```
